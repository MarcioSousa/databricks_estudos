#01 - Databricks Workspace Walkthrough
Neste passo a passo, abordaremos vários componentes essenciais do espaço de trabalho do Databricks, incluindo:
1. **Explode the Databricks Workspace Homepage** e entenda sua utilidade para navegação de alto nível.
2. **Utilize o painel de navegação** para alternar entre os diversos recursos do Databricks.
3. **Inspecionar e alterar as configurações do usuário**.
4. **Configure a classroom Setup** para todas as demonstrações e laboratórios.
5. **Navegar pelos ativos e objetos do Workspace**, veremos como se mover de forma eficiente pelas diferentes áreas da interface do Databricks.
6. **Trabalhando com notebooks**, vamos abordar como criar, editar e executar código em notebooks.
7. **Unity Catalog e Catalog Explorer**, abordando como atribuir e gerenciar permissões para diferentes usuários e grupos, garantindo acesso seguro aos dados, além de como localizar diversos ativos registrados no Catálogo do Unity.
8. **(Opcional) Computação**, explicando como usar a interface do usuário para criar um novo cluster.
9. **(Opcional) Funcionalidade Git**, onde discutiremos como conectar-se a um repositório Git e criar pastas Git.

_Observação: Haverá muito pouca programação nesta demonstração._

##1.Expanda a página inicial do Databricks Workspace
1. Se você clicar no logotipo do Databricks no canto superior esquerdo, será direcionado para uma tela que diz **Welcome to Databricks**.
2. Você verá cinco abas diferentes com os seguintes rótulos:
    * **Recentes (Recents)**
    * **Favoritos (Favorites)**
    * **Popular (Popular)**
    * **Mosaic AI**
    * **O que há de novo (What's New)**
3. Em **Recentes (Recents)**, você encontrará todos os seus notebooks que foram abertos ou editados recentemente.
4. Em **Favoritos (Favorites)**, sempre que você adicionar um notebooks, tabela ou qualquer outro tipo de recurso do Databricks aos seus favoritos, ele aparecerá aqui. Para adicionar um item aos favoritos, selecione o ícone de estrela que fica próximo ao nome do objeto.
5. Em **Popular (Popular)**, você encontrará tabelas, notebooks e outros recursos populares no espaço de trabalho. São objetos aos quais você tem acesso e que seus colegas utilizaram com frequência suficiente nos últimos 30 dias.
6. Em **Mosaic AI**, você verá os modelos adicionados recentemente e os modelos em destaque registrados no Mosaic AI Model Serving.
7. Em **O que há de novo (What's New)**, você pode ver os anúncios e atualizações recentes sobre a plataforma. Nessa tela, você pode ler mais sobre o recurso, experimentá-lo no ambiente ou criar um objeto diretamente a partir do anúncio.
8. Além disso, você tem a opção de **Busca (Search)** na parte superior da tela, onde pode pesquisar dados, notebooks ou outros objetos e recursos dentro da plataforma.

##2. Utilize the navigation panel
1. Click on the **hamburguer icon** (three horizontal lines stacked on top of each other) at the top left. Clicking on it will reveal or hide the left sidebar navigator. When it is hidden, you can reveal the navigation panel by hovering over the hamburger icon.

2. In the left sidebar, you will find the following items at the top:
    * **+ New**
    * **Workspace**
    * **Recents**
    * **Catalog**
    * **Workflows**
    * **Compute**
    * **Marketplace**

3. If you click on **+ New** at the very top of the sidebar navigation menu, you will have options to create a new:
    * **Notebook**
    * **Query**
    * **Dashboard**
    * **Genie space**
    * **Job**
    * **DLT Pipeline**
    * **Alert**
    * **Experiment**
    * **Model**
    * **App**
    * At the very bottom, you will see **More**, which allows you to create with:
        * **Git folder**
        * **Cluster**
        * **Sql Warehouse**
        * **Serving Endpoint**
4. You will also find a grouping of SQL menus:
    * **Sql Editor**
    * **Queries**
    * **Dashboards**
    * **Features**
    * **Models**
    * **Serving**

###3. Inspect and change user settings
1. At the top right, you can click on the user icon, typically a letter associated with your user name, and select **Settings**.
2. In **Settings**, you will see **User Settings** with five diferent options:
    * **Profile**
    * **Preferences**
    * **Developer**
    * **Linked Accounts**
    * **Notifications**  
Note that if you have administrative proviledges, you will have an additional set of menu options for **Workspace admin**. Those options are not covered in this content.
3. If you click **Profile**, you will see:
    * Your display name
    * The group(s) you belong to
4. Under **Preferences**, you will see two options to manage:
    * **General: Language**
    * **Appearance: Interface Theme**  
You can click on these options if you wish to change them.  
5. Under **Developer**, you will see various options as you scroll down, including
    * Managing **Access tokens**
    * **SQL Query Snippets**
    * **Editor Settings**
      * **General** settings
      * **Code editor** settings
      * **Experimental features** toggles
You'll see various options toggled on or off as you scroll down.
6. Under Linked Account, you will find information about Git integration. Here, you can:
    * Link a Git account
    * Select a **Pesonal Access Token** for a given Git provider, such as GitHub.
7. Finally, under Notifications, you can manager when you'll be notifield for:
    * **Model Registry events**
    * **Comment email notifications**
    * **Account-level email communications**
    * **Promotional email communications**
