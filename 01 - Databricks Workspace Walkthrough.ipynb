#01 - Databricks Workspace Walkthrough
In this walkthrough, we will be covering several key components of the Databricks workspace, including:
1. **Explode the Databricks Workspace Homepage** and understand its usefulness for high-level navigation.
2. **Utilize the navigation panel** to switch between various Databricks features.
3. **Inspect and change user settings**.
4. **Execute the classroom Setup** for all the demos and labs.
5. **Navigate Workspace assets and objects**, where we will look into how to efficiently move through different areas of the Databricks interface.
6. **Work with notebooks**, where we will go over how to create, edit, and run code in notebooks.
7. **Unity Catalog and Catalog Explorer**, covering how to assign and manage permissions for different users and groups to ensure secure access to data in addition to how to locate various assets registered to Unity Catalog.
8. **(Optional) Compute**, going over how to use the UI to create a new cluster.
9. **(Optional) Git Functionality**, where we'll discuss how to connect to a git repository and create git folders.

_Note: There will be very little coding in this demostration._

##1.Explode the Databricks Workspace Homepage
1. If you click on the **Databricks logo** at the top left, you'll be taken to a screen that says **Welcome to Databricks**.
2. You will see five different tabs labeled:
    * **Recents**
    * **Favorites**
    * **Popular**
    * **Mosaic AI**
    * **What's New**

3. Under **Recents**, you'll find all of your notebooks that have been recently opened or worked on.
4. Under **Favorites**, anytime you favorite a notebook, table, or any kind of asset within Databricks, you will find it here. To Favorite an item, you'll select the Star icon that will be near objects name.
5. Under **Popular**, this is where you can discover popular tables, notebooks, and other assets within the workspace. Theses are objects that you have access to that your coworkers have engaged with frequently enough over the last 30 days.
6. Under **Mosaic AI**, you will see newly added and featured models registered with Mosaic AI Model Serving.
7. Under **What's New**, you can see recent announcements and updates about the platform. Form this screen, you can read more about the feature, check it out within the environment, or create an object directly from the announcement.
8. Additionally, you have the **Search** at the top of the screen where you can search for data, notebooks, or other objects and assets within the platform.
